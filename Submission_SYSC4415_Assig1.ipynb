>a.   What is the expected value from this sample? Using an unbiased estimator, what is the sample variance and standard deviation?

Expected Value:
$$E(X) = \mu = \frac{1}{n} * \Sigma_{i=1}^{n}x_{i} = \frac{1}{10}(2+1+4+1+3+2+3+2+5+1) =  2.4 $$

Sample Variance:
$$\sigma^2 = E(X-\mu)^2 = Pr(X=x_{1})(x_{1}-\mu)^2 + Pr(X=x_{2})(x_{2}-\mu)^2 + Pr(X=x_{3})(x_{3}-\mu)^2 + Pr(X=x_{4})(x_{4}-\mu)^2 + Pr(X=x_{5})(x_{5}-\mu)^2 + Pr(X=x_{6})(x_{6}-\mu)^2 + Pr(X=x_{7})(x_{7}-\mu)^2 + Pr(X=x_{8})(x_{8}-\mu)^2 + Pr(X=x_{9})(x_{9}-\mu)^2 + Pr(X=x_{10})(x_{10}-\mu)^2 = (2)(2-2.4)^2 + (1)(1-2.4)^2 + (4)(4-2.4)^2 + (1)(1-2.4)^2 + (3)(3-2.4)^2 + (2)(2-2.4)^2 + (3)(3-2.4)^2 + (2)(2-2.4)^2 + (5)(5-2.4)^2 + (1)(1-2.4)^2 = 53.04$$

Standard Deviation:
$$\sigma = \sqrt{\sigma^2} = \sqrt(53.04) = 7.28$$

$$\sigma^2 = E[(x-\mu)^2] = Pr(X=x_1)(x_1-\mu)^2 + Pr(X=x_2)(x_2-\mu)^2 + Pr(X=x_3)(x_3-\mu)^2 +  Pr(X=x_4)(x_4-\mu)^2 + Pr(X=x_5)(x_5-\mu)^2$$

>d.   Find the probability that a student slept 2 hours the night before ($x=2$), given that they did not need excess amounts of coffee. That is, find $Pr(2|-)$. Hint: $Pr(+)$ can be found by summing over $Pr(+|x)Pr(x)$, for all $x$. $Pr(-|x)$ can be derived from $Pr(+|x)$.

Calculating Pr(+):

$$Pr(+) = \Sigma_{i=1}^{n}(Pr(+|x_i)*Pr(x_i) = (0.25)(0.75) + (0.30)(0.55) + (0.20)(0.35) + (0.10)(0.20) + (0.15)(0.10) = 0.4575 $$

Calculating Pr(-):
$$ 1 = Pr(+) + Pr(-) $$
$$ Pr(-) = 1 - Pr(+) = 1 - 0.4575 = 0.5425 $$

Calculating Pr(-|x=2):
$$ 1 = Pr(+|x=2) + Pr(-|x=2) $$
$$ Pr(-|x=2) = 1 - Pr(+|x=2) = 1 - 0.55 = 0.45 $$

Calculating  Pr(x=2|-):
$$ Pr(x=2|-) = \frac{Pr(-|x=2)Pr(X=2)}{Pr(-)} = \frac{0.45*0.30}{0.5425} = 0.249 $$

Therefore the probability that the student slept 2 hours, given that the student was not given excess amount of coffee is 0.249

## Step 1: Load the CSV file, Plot the data, and Calculate the Least Squares solution for Linear Regression

```python
#Load data
url="https://raw.githubusercontent.com/jrgreen7/SYSC4906/master/F2021/Assignments/Assig1/Assig1Q4.csv"
data = pd.read_csv(url, header=None)
X_data = np.array(data[0])
Y_data = np.array(data[1])

#Plot the x and y values
plt.scatter(X_data, Y_data)
plt.show()
```

## Step 2: Linear model $y=mx+b$

```python
# Augment the x vector
X = np.ones(shape=(X_data.size, 2))
for i in range(X_data.size):
  X[i][0] = X_data[i]

# Compute beta
B = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), Y_data)
print(f"From Beta, w={B[0]} and b={B[1]}")

#Plot the line of best fit y=wx+b
plt.title("Linear regression")
plt.plot(X_data, Y_data, 'o')
plt.plot(X_data, B[0]*X_data, B[1])
plt.xlabel("X")
plt.ylabel("Y")

# Compute the MSE
MSE = np.sum((Y_data - np.dot(X,B))**2) / X_data.size
print(f"MSE = {MSE}")
```

## Step 3: Cubic model $y=ax^3+bx^2+cx+d$

```python
# Augment the x vector
cubicX = np.ones(shape=(X_data.size, 4))
for i in range(X_data.size):
  cubicX[i][1] = X_data[i]
  cubicX[i][2] = X_data[i] ** 2
  cubicX[i][3] = X_data[i] ** 3



# Compute beta
cubicB = np.matmul(np.matmul(np.linalg.inv(np.matmul(cubicX.transpose(),cubicX)),cubicX.transpose()),Y_data)

print(f"From Beta, bias={cubicB[0]} and weight={cubicB[1]} second weight={cubicB[2]} third weight={cubicB[3]}")

#Used for spacing
xlin = np.linspace(X_data.min(), X_data.max(), 100000)

#y=ax^3+bx^2 +cx+d
cubicY = cubicB[3] * xlin**3 + cubicB[2] * xlin**2 +  cubicB[1] * xlin + cubicB[0]

#Plot the line of best fit y=wx+b
plt.title("Cubic model")
plt.scatter(X_data, Y_data)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(xlin, cubicY)

# Compute the MSE
MSE_cubic = np.sum((Y_data - np.dot(cubicX,cubicB))**2) / X_data.size
print(f"MSE = {MSE_cubic}")
```
